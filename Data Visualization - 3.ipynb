# Seaborn

* Seaborn is a Python data visualization library based on matplotlib. 
* It provides a high-level interface for drawing attractive and informative statistical graphics.
* Seaborn will provide a variety of visualization patterns than matplotlib.

```python
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
```

## Available Datasets

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

# Different Plots

## Barplot

```python
sns.barplot(x = tips['day'],y=tips['tip'])
```

**Input data can be passed in a variety of formats, including: **

    - Vectors of data represented as lists, numpy arrays, or pandas Series
      objects passed directly to the ``x``, ``y``, and/or ``hue`` parameters.

```python
sns.barplot(x = 'day',y='tip',data = tips)
```

```python
sns.barplot(x = 'day',y='tip',data = tips , color='blue')
```

```python
sns.barplot(x = 'day',y='tip',data = tips , color='blue',saturation=0.3)
```

```python
sns.barplot(x = 'day',y='total_bill',data = tips)
```

```python
sns.barplot(x = 'day',y='tip',data = tips,hue='sex')
```

## Heatmap

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights = flights.pivot('month','year','passengers')
```

```python
flights.head()
```

```python
sns.heatmap(flights)
```

```python
sns.heatmap(flights,annot=True,fmt='d',linewidths=0.8)
```

## Clustermap

```python
sns.clustermap(flights)
```

```python
sns.clustermap(flights,linewidth = 1)
```

```python
sns.clustermap(flights,z_score=0)
```

## Regplot

```python
sns.regplot(x='total_bill',y='tip',data=tips,color='green')
```

```python
sns.regplot(x='total_bill',y='tip',data=tips,scatter_kws={'color':'red'},line_kws={'color':'black'})
```
