# Data Visualization

Data Visualization is a presentation of the data in a pictorial format or graphical format.

# Matplotlib

* Excellent Data Visualization Library
* Multi-Platform Support
* It supports several plots like line, bar, scatter, histogram etc.

## Basic Elements of the plot
![](https://matplotlib.org/3.1.1/_images/anatomy.png)

```python
import matplotlib
```

```python
from matplotlib import pyplot as plt 
import numpy as np
```

```python
y = [2,4,6,8,10] 
```

```python
plt.plot(y)
plt.show()
```

** Title **

** xlabel , ylabel **

```python
plt.plot(y)
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.show()
```

```python
# subplot (row,col,fig_num)
plt.subplot(1,2,1)
plt.plot(y)
plt.subplot(1,2,2)
plt.plot(y)
plt.show()
```

## Annotating a plot

```python
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
plt.plot(t, s)

plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black'),
            )
plt.ylim(-2,2)
plt.show()
```

## Different Plots

## Line Plot

```python
# x-axis values 
x = [1,2,3,4,5] 
  
# Y-axis values 
y = [2,4,6,8,10] 
  
# Function to plot 
plt.plot(x,y) 
  
# function to show the plot 
plt.show() 
```

## Bar plot 

```python
# x-axis values 
x = [5, 2, 9, 4, 7] 
  
# Y-axis values 
y = [10, 5, 8, 4, 2] 
  
# Function to plot the bar 
plt.bar(x,y) 

#labels = ['a','b','c','d','e']
#plt.text()
# function to show the plot 
plt.show() 
```

## Histogram

```python
# Y-axis values 
y = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100] 
  
# Function to plot histogram 
plt.hist(y,bins = 5) 
  
# Function to show the plot 
plt.show() 
```

## Scatter Plot

```python
# x-axis values 
x = [5, 2, 9, 4, 7] 
  
# Y-axis values 
y = [10, 5, 8, 4, 2] 
  
# Function to plot scatter 
plt.scatter(x, y) 
  
# function to show the plot 
plt.show() 
```

## Area plot

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 1, 2, 3, 5]
y2 = [0, 4, 2, 6, 8]
y3 = [1, 3, 5, 7, 9]
plt.stackplot(x,y1,y2,y3)
```

# Pie Chart

```python
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.4, 0, 0) 
```

```python
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')
plt.show()
```

## Contour 

```python
xlist = np.linspace(-3.0, 3.0, 100)
ylist = np.linspace(-3.0, 3.0, 100)
```

```python
xlist[:5]
```

```python
ylist[:5]
```

```python
X, Y = np.meshgrid(xlist, ylist)
```

```python
X.shape,Y.shape
```

```python
X
```

```python
Y
```

```python
Z = np.sqrt(X ** 2 + Y ** 2 )
Z
```

```python
Z.shape
```

```python
levels = [0.0, 0.2, 0.5, 0.9, 1.5, 2.5, 3.5]
contour = plt.contour(X, Y, Z)
plt.clabel(contour)
contour_filled = plt.contourf(X, Y, Z, levels)
plt.colorbar(contour_filled)
plt.title('Plot from level list')
plt.xlabel('x (cm)')
plt.ylabel('y (cm)')
plt.show()
```
